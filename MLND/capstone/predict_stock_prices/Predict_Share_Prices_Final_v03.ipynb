### Split Data into Training, Validation and Test Datasets

```python
def splitDataset(input_dataset):
    data_values = input_dataset.values
    data_size = len(input_dataset)

    train_test_split = int(np.ceil(8*data_size/float(10)))  

    test_dataset = data_values[train_test_split:,:]
    #y_test = y[train_test_split+100:]

    train_valid_split = train_test_split - (data_size - train_test_split)

    train_dataset = data_values[:train_valid_split,:]
    #y_train = y[:train_valid_split]

    valid_dataset = data_values[train_valid_split:train_test_split,:]
    #y_valid = y[train_valid_split:train_test_split]
    return train_dataset, valid_dataset, test_dataset
    
```

```python
data_SP500_train, data_SP500_valid, data_SP500_test = splitDataset(normalized_data_SP500_AdjClose)

data_IBM_train, data_IBM_valid, data_IBM_test = splitDataset(normalized_data_IBM_AdjClose)
data_MSFT_train, data_MSFT_valid, data_MSFT_test = splitDataset(normalized_data_MSFT_AdjClose)
data_PFE_train, data_PFE_valid, data_PFE_test = splitDataset(normalized_data_PFE_AdjClose)
data_WMT_train, data_WMT_valid, data_WMT_test = splitDataset(normalized_data_WMT_AdjClose)
data_XOM_train, data_XOM_valid, data_XOM_test = splitDataset(normalized_data_XOM_AdjClose)
data_MYL_train, data_MYL_valid, data_MYL_test = splitDataset(normalized_data_MYL_AdjClose)
data_AAPL_train, data_AAPL_valid, data_AAPL_test = splitDataset(normalized_data_AAPL_AdjClose)
data_GOOG_train, data_GOOG_valid, data_GOOG_test = splitDataset(normalized_data_GOOG_AdjClose)
```

## Tranform Data to window dependency supervised data

```python
def window_transform_series(series, window_size):
    # containers for input/output pairs
    X = []
    y = []
    
    numWindows = len(series) - window_size
    
    for i in range(numWindows):
        lst = series[i:i + window_size]
        out = series[i + window_size]
        X.append(lst)        
        y.append(out)       
        
        
    # reshape each 
    X = np.asarray(X)
    X.shape = (np.shape(X)[0:2])
    y = np.asarray(y)
    y.shape = (len(y),1)
    
    return X,y 
```

### Model Definition

```python
np.random.seed(1234)

model = Sequential()

model.add(LSTM(input_shape=(None,1),units=50, return_sequences=True))
model.add(Dropout(0.1))

model.add(LSTM(256,return_sequences=False))
model.add(Dropout(0.1))

model.add(Dense(units=1))
model.add(Activation('linear'))

optimizer = keras.optimizers.Adagrad(lr=0.005, epsilon=1e-08, decay=0.0)

model.compile(loss='mse', optimizer=optimizer)
```

```python
class LossHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.losses = []
        self.val_losses = []
        #self.rmse = []
        #self.val_rmse = []

    def on_epoch_end(self, batch, logs={}):
        self.losses.append(logs.get('loss'))
        self.val_losses.append(logs.get('val_loss'))
        
    def on_train_end(self, logs={}):
        plt.plot(self.losses, color='orange')
        plt.plot(self.val_losses, color='blue')
        plt.title("Losses vs Epochs")
        plt.ylabel("mse loss")
        plt.xlabel("epoch")
        plt.legend(['Training Loss','Validation Loss'],loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
history = LossHistory()
```

```python
filepath="Predict_Share_Prices_model_weights_best.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=True, mode='min')
callbacks_list = [checkpoint,history,EarlyStopping(monitor='val_loss', patience=20)]
```

## Conclusion

#### Prediction of future stock prices has been quite challenging and tedious task. After spending lot of time on framing the data for supervised learning, then model selection and tuning, it has been possible to build a model performing close( rather better) to the model in practice in industry. The Deep Learning LSTM model has worked than the benchmarked models.
#### Feature selection showed that there is enough information in previous Adj Close prices to predict the future values. Using other features such as High, Low , Open, Close and calculated features DailyReturn did not add any value to the model.
#### Different window sizes ( previous days values) experimentation also showed that 5 days data carried enough information to predict future values and adding additional days data did not add to the performance.
#### Batch sizes for model training also affected performance and although using larger batch sizes made the model run faster, but it resulted in low model accuracy.
#### The problem of predicting stock prices actually involves identifying patterns in sequences and LSTMs are designed to predict sequence patterns. LSTMs have been successfully used in identifying patterns in images, speech recognition and some attempt has been made to predict future time series. The results from this project show that LSTMs can be used successfully in predicting future prices.
